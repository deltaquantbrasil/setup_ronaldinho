> # **BACKTEST** - Estratégia *Ronaldinho*
---

A estratégia é bem simples:

* Timeframe de 5 ou 10 minutos;
* Esperar o candle de 9:55 ou 9:50 fechar e terá que ser um candle de baixa;
* Comprar na abertura do candle das 10 horas com stop de 300 pontos;
* Saída da operação no fechamento do candle de 10 horas ou com alvo fixo ou por tempo.

_OBS.: estratégia retirada do canal do **Viana Trader** no vídeo: [OPERANDO COM RONALDINHO NO MINI ÍNDICE + MINI DÓLAR](https://www.youtube.com/watch?v=-cUPbGoWzH0)_

_OBS1.: a estratégia foi alterada para tentar melhorar o resultado._

# 1 - Bibliotecas

```python
# Análise e manipulação de dados gerais
import numpy as np
import pandas as pd

# Dados séries temporais
from datetime import datetime, timedelta
import pytz

# Dados OHLCV
from mt5linux import MetaTrader5
from datetime import datetime

# connecto to the server
mt5 = MetaTrader5()
mt5.execute('import datetime')

# Bibliotecas gráficas
import mplfinance as mpf
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

# 2 - Obtenção dos dados

```python
mt5.initialize()
```

```python
ticker = 'WIN$N'
```

```python
timezone = pytz.timezone("America/Sao_Paulo")
data_inicio = datetime.now() - timedelta(days=300)
data_fim = datetime.now()
```

```python
dados_ohlcv = mt5.copy_rates_range(ticker, mt5.TIMEFRAME_M1, data_inicio, data_fim)
dados_ohlcv = pd.DataFrame(dados_ohlcv)
dados_ohlcv['ticker'] = str(ticker)
dados_ohlcv['timeframe'] = '01_min'
dados_ohlcv.index = pd.to_datetime(dados_ohlcv['time'], unit='s')
dados_ohlcv = dados_ohlcv.drop('time', axis=1)
dados_ohlcv
```

## 2.1 - Fazendo o resample dos dados

```python
df = dados_ohlcv.resample('5T').agg({'open': 'first', 'high': 'max', 'low': 'min', 'close': 'last', 'real_volume':'sum', 'ticker': 'first'})
df.tail()
```

## 2.2 - Definindo a direção do candle

```python
df.loc[df['close'] > df['open'], 'direction'] = 1
df.loc[df['close'] < df['open'], 'direction'] = -1
df['direction'].fillna(0)
df.dropna(inplace=True)
```

# 3. Filtrando as entradas

```python
def entrada_compra(hora_minuto, direcao, hora):
    regra_1 = (hora_minuto == hora)
    regra_2 = (direcao == -1.0)
    
    return (regra_1 & regra_2)
```

```python
# Definindo o stop da operação
alvo = 100
stop = 350
```

```python
data = df.copy()

data['hour'] = data.index.strftime('%H:%M')
data['compra'] = entrada_compra(data.hour, data.direction, '09:55')
data.tail()
```

### Filtrando as entradas de compra

```python
df_compra = data[data.compra == True]
df_compra
```

### Conferindo o ponto de entrada no gráfico

```python
n_index = 0
inicio = df_compra.iloc[n_index].name - timedelta(minutes=15)
fim = df_compra.iloc[n_index].name + timedelta(minutes=60)

candle_index = 4

dfpl = df.loc[inicio:fim]

candle_signal = dfpl.index[candle_index-1].strftime('%Y-%m-%d %H:%M:%S')
candle_entry = dfpl.index[candle_index].strftime('%Y-%m-%d %H:%M:%S')
candle_end = dfpl.index[-1].strftime('%Y-%m-%d %H:%M:%S')

ponto_entrada = dfpl['open'].loc[candle_entry]
ponto_stop = dfpl['open'].loc[candle_entry] - stop
ponto_take = dfpl['open'].loc[candle_entry] + alvo

lines = [[(candle_entry,ponto_stop),(candle_end, ponto_stop)],
         [(candle_entry,ponto_entrada),(candle_end, ponto_entrada)],
         [(candle_entry,ponto_take),(candle_end, ponto_take)]]

alines = dict(alines=lines, linestyle='-.', colors=['tab:red', 'tab:blue', 'tab:green'], alpha=0.9, linewidths=1.5)
vlines=dict(vlines=candle_signal,linewidths=20, alpha=0.3, colors='tab:gray')
mpf.plot(dfpl, title='Mini Índice - Ronaldinho', type='candle', style='classic', vlines=vlines, alines=alines)
```

## 4. Backtest Raiz

```python
stop = 345
take = 105

position = 0

ls_data_compra_entrada = []
ls_preco_compra_entrada = []

ls_data_compra_saida = []
ls_preco_compra_saida = []
ls_motivo_compra_saida = []

for i in range(len(data)):
    if (position == 0) and (data['compra'].iloc[i] == True):
        ls_data_compra_entrada.append(data.index[i+1])
        ls_preco_compra_entrada.append(data['open'].iloc[i+1])
        position = 1
        
        preco_entrada = data['open'].iloc[i+1]
        stop_loss = preco_entrada - stop
        take_profit = preco_entrada + take
        
    elif (position == 1) and ((data['low'].iloc[i] <= stop_loss) or (data.index[i].hour >= 17)):
        ls_data_compra_saida.append(data.index[i])
        ls_preco_compra_saida.append(stop_loss)
        ls_motivo_compra_saida.append('STOP')
        position = 0
        
    elif (position == 1) and ((data['high'].iloc[i] >= take_profit) or (data.index[i].hour >= 17)):
        ls_data_compra_saida.append(data.index[i])
        ls_preco_compra_saida.append(take_profit)
        ls_motivo_compra_saida.append('TAKE')
        position = 0
        
    elif (position == 1) and (data['hour'].iloc[i] == '10:10'):
        ls_data_compra_saida.append(data.index[i])
        ls_preco_compra_saida.append(data['open'].iloc[i])
        ls_motivo_compra_saida.append('TIME')
        position = 0
        
        
tabela_result_compras = pd.DataFrame(zip(ls_data_compra_entrada, ls_preco_compra_entrada, ls_data_compra_saida, ls_preco_compra_saida, ls_motivo_compra_saida),
                                    columns=['data_entrada', 'preco_entrada', 'data_saida', 'preco_saida', 'motivo_saida'])
tabela_result_compras['data_entrada'] = pd.to_datetime(tabela_result_compras['data_entrada'], errors='ignore')
tabela_result_compras['data_saida'] = pd.to_datetime(tabela_result_compras['data_saida'], errors='ignore')
tabela_result_compras['resultado'] = tabela_result_compras['preco_saida']/tabela_result_compras['preco_entrada'] - 1
tabela_result_compras['resultado_pontos'] = tabela_result_compras['preco_saida'] - tabela_result_compras['preco_entrada']
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(14,6))
plt.title('Pontos acumulados da Estratégia')
plt.plot(tabela_result_compras['resultado_pontos'].cumsum(), lw=0.7, color='black', label='Resultado em pontos')
plt.axhline(0, color='gray', ls='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show();
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(14,6))
plt.title('Porcentagem acumulada da Estratégia')
plt.plot(tabela_result_compras['resultado'].cumsum()*100, lw=0.7, color='black', label='Resultado acumulado')
plt.axhline(0, color='gray', ls='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show();
```

```python
soma_take = tabela_result_compras[tabela_result_compras['motivo_saida'] == 'TAKE']['resultado_pontos'].sum()
soma_stop = tabela_result_compras[tabela_result_compras['motivo_saida'] == 'STOP']['resultado_pontos'].sum()
soma_time = tabela_result_compras[tabela_result_compras['motivo_saida'] == 'TIME']['resultado_pontos'].sum()

qtd_take = tabela_result_compras[tabela_result_compras['motivo_saida'] == 'TAKE']['resultado_pontos'].count()
qtd_stop = tabela_result_compras[tabela_result_compras['motivo_saida'] == 'STOP']['resultado_pontos'].count()
qtd_time = tabela_result_compras[tabela_result_compras['motivo_saida'] == 'TIME']['resultado_pontos'].count()

print('\n\nBacktest do RONALDINHO')

print(f'\nInício do backtest: {data.index[0]}')
print(f'Fim do backtest: {data.index[-1]}')
print(f'Quantidade de dias: {data.index[-1] - data.index[0]}')

print(f'\nQuantidade de saídas no take profit: {qtd_take}')
print(f'Quantidade de saídas no stop loss: {qtd_stop}')
print(f'Quantidade de saídas no tempo: {qtd_time}')

print(f'\nSoma dos pontos na saída por take profit: {soma_take}')
print(f'Soma dos pontos na saída por stop loss: {soma_stop}')
print(f'Soma dos pontos na saída por tempo: {soma_time}')

print(f'\nResultado final: {soma_take + soma_stop + soma_time} pontos')
```
